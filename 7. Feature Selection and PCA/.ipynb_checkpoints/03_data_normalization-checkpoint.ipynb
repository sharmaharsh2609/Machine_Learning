## Normalization/ Standardization
- One of the step that we do in data-preprocessing.
- Normalizing data means shifting the data to the mean=0 and std=1.

```python
import matplotlib.pyplot as plt, numpy as np
import pandas as pd
```

### Why do we need Normalization
- We normalize the data to change the values of all the columns to a common scale. eg- if we have 5features in X_train, and all of them has differnet mean and std, then by normalizing data, we shift every column to a common scale i.e- 0-mean and unit-std.
- By normalizing data, our model works lot faster.
- And by normalizing(shifting columns) does not affect the predictions bcoz we only change their scale. But data is still same.

```python
plt.style.use("seaborn")
```

```python
dfx = pd.read_csv("linearX.csv")
dfy = pd.read_csv("linearY.csv")

x = dfx.values[:, -1]  # choose x as last column of xdata only
y = dfy.values[:, -1]  # choose y as last column of ydata only

plt.scatter(x,y)
plt.show()
```

- This data has y_values smaller than 1.00 and x is in range 6-16. 

```python
u = x.mean()
std = x.std()

x_new = (x-u)/std

plt.scatter(x,y, label="Raw data initially")
plt.scatter(x_new, y, color="orange",  label="Normalized Data")
plt.legend()  # to create a legend box
plt.show()
```

- We can see our new data is centered around mean=0. Here our new_x has a mean of 0(orange color data)

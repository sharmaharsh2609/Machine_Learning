# K-Means
- Common clustering technique, helps to solve many problems efficiently.
- K-Means is an example of Hard Clustering, where every point belongs only to one cluster.

## Motivation Problem - *Pizza Parlour Study*

**Problem Statement** Given the list of locations of customers who frequently order Pizza we want to find out optimal locations of Pizza Parlours where they should be opened.

## K-Means Algorithm

**Input Data** : X = {x1,...,xN}

**Goal** : Partition the data among some K number of clusters. Let us assume K is known to us.

Let $µ_k$ denote the center of Kth Cluster ($u_k$ will be vector)

So we need to ﬁnd an **assignment of data points to clusters**,
as well as a **set of cluster centers{µk**}, such that the sum of the squares 
of the distances of each data point to its closest cluster center µk, is a minimum. 

We have to minimize the error: sum of distance between point and it's nearest cluster.

- Then we do it like: (2x-1), x$\in${0.0,1.0}. Thus (2x-1) lies in range 2*{0,1}-1 = {0,2}-1 = {-1,1}. Like this we can generate any random number in range -1 to 1.

```python
k = 5  # here we have 5clusters. So pass k=5
color  = ["green","red","blue","yellow","orange"]  # we will use these colors for above 5 clusters

clusters = {}  # initialize an empty dictionary to store all k- clusters
## add all the clusters in each iteration to this dictionary.

for i in range(k):  # iterate over all clusters.
    ## do following operations for all k-clusters in k- iterations 
    
    ## see in cell above to this one to understand meaning of this line code:
    center = 10*(2*np.random.random((X.shape[1],))-1) # generate random center for ith cluster
    points = [] 
    # for each ith cluster, we need a list of all the points belonging to ith cluster
    
    ## create a new sub_dictionary for each ith cluster and add center point to it, list of points
    ## and color from the list that we defined initially before this loop.
    sub_cluster = {
        'center':center,
        'points':points,
        'color': color[i]
        }
    
    ## we will create a sub_dictionary for each cluster and add that to our main dictionary
    ## take key as index and value as sub_dictionary. Like this we will have 5 dictionaries for 
    ## 5 different clusters
    clusters[i] = sub_cluster  # this is a dictionary of dictionaries.
    ## see below where we print this clusters dictionary. See how it looks like
```

- We can see: We have 5 clusters where each cluster has it's own centerr, points and color. 
- For now, we have only assigned some random k-points to it. But we have not assignes any points to any clusters. So points lists are empty for each cluster.
- To distribute all data points in these k-clusters, we need to calculate distance of every point from all k-clusters and then after sorting, we will assign each point to the cluster which has minimum distance from it.

```python
## this is same function that we defined in knn algo. It calculates distance b/w 2points
## where both points are vectors with some features(For this example: we have 2 features)
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
# Assign every data point to one of the clusters
def assignPointToClusters(clusters): #E-Step
    ## iterate over all the points in X
    for ix in range(X.shape[0]):  
        dist = [] # initialize an empty list
        curr_x = X[ix]    # this is current point.

        # for every point: iterate over all k-clusters to find distance of cur_pt from all clusters
        for kx in range(k):
            d = distance(curr_x,clusters[kx]['center'])
            dist.append(d)
        ## now 'dist' stores distance of all clusters from current point.
        ## so get index of cluster which have minimum distance from current point.
        min_idx = np.argmin(dist)     
## clusters[min_idx] gives us cluster with minimum point. 
## Now in that cluster append cur_x/cur_point to the points_list of that cluster
        clusters[min_idx]['points'].append(curr_x)
```

- E-step and M-step means Expectation and Maximization. These terms come from EM algorithm. EM algo is a seperate algo. EM algo is generalization of k-means. or we can say: **k-means is a special case of EM algorithm**

```python
# M-Step -> We update every cluster center according to the mean of the points in it
def updateClusters(clusters): 
    for kx in range(k):
        pts = np.array(clusters[kx]['points']) # convert points_list for each cluster to np.array
        
        if pts.shape[0]>0: #If a cluster has some-nonzero points, then assign new center
        ## find mean along axis=0: means find mean of all points for 1st col and for 2nd col
        ## this gives us (ux,uy) mean of all x and y values of all points 
        ## and this is our uk: "uk=(ux,uy)"  : mean for kth cluster.
            new_u = pts.mean(axis=0)
            clusters[kx]['center'] = new_u   ## update center as new_mean
            clusters[kx]['points'] = [] # we need to clear the list for next E-step.
            ## E-step will now divide all points in k-clusters
                    
```

- We will need to write code again and again to plot these points and clusters. So we will create a function to plot and call this function whenever we want to plot points and clusters. 

```python
def plotClusters(clusters):
    
    for kx in range(k):  ## iterate over all points
        ## convert list of points of each cluster to numpy array in each iteration
        pts = np.array(clusters[kx]['points'])
        
        # Plot the points ( use try block bcoz some clusters may have 0 points in there 'points'
        # list. We can't plot 0 point. So use try except to handle that error)
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
        except:
            pass

        #Plot the cluster center
        uk = clusters[kx]['center']   # center points of each cluster in ith iteration
        plt.scatter(uk[0],uk[1],color="black",marker="*")  ## plot center point with *
```

- Initially these are all the clusters and we have not assigned any points to any clusters. So our plotClusters() function will not show any points.
- To show points: we need to call assignPointToClusters() function first, then call plotCluster() function to plot.

```python
assignPointToClusters(clusters)  # pass dictionary of clusters that we defined in the beginning.
for i in range(len(clusters)):
    print(len(clusters[i]['points']))  # see number of points assigned to each point
```

- This is E-step. Now all the points are distributed to all the k-clusters(to 5clusters here). 

### Note: If any of these cluster have 0 points assigned to it, then execute all cells again.(or reassign random points to all clusters)
- This is bcoz if any of these have 0 points then that will stay far away and all other points will divide points to them. This is the problem with k-means Clustering. We will se it later.

```python
plt.figure(figsize=(6,4))
plotClusters(clusters)   # use the function we created earlier to plot all points and clusters.
```

- This is the case in the beginning.
- Now update all k-centers as mean of all the points assigned to them according to M-step

```python
## update k-center points
updateClusters(clusters) # M-step
```

- In updation, all the data in points list is erased. And points lists for all cluster are empty now again.
- After Updation again call E-step: assignPointCluster() to again assign all the points and plot the points to see how it looks like. And then perform M-step for updation. 

- We can write this below code and execute a cell it multiple lines. But I'm using here subplots to illustrate this algo stepwise:

```python
## Repeat this step for a few number of times:()
plt.figure(figsize=(16,10))
for i in range(12):
    plt.subplot(3,3,1+i)
    assignPointToClusters(clusters)
    plotClusters(clusters)
    updateClusters(clusters)
plt.show()
```

- We can see, that it converges to minimum value after updating for only 3times. 

### **k-means sometimes gives us only 3 or 4 clusters in above case. This is bcoz k-means get stuck on local minima sometimes. So random initialization matters a lot in k-means**

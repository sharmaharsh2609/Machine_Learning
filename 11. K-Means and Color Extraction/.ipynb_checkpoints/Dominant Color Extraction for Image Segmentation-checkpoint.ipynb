## Dominant Color Extraction for Image Segmentation
- **Image Segmentation**
- Segmentation partitions an image into regions having similar visual appearance corresponding to parts of objects
- We will try to extract the most dominant 'K' Colors using K-Means
- We can apply K-Means with each pixel will reassigned to the closest of the K Colors, leading to segmentation

```python
import matplotlib.pyplot as plt
import cv2
```

```python
im =  cv2.imread('elephant.jpg') #Reads an image into BGR Format

im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
original_shape = im.shape    ## required in later part at end
print(im.shape)
```

- It is a (300x500) image with 3channels or we can say depth of image=3. And cv2 reads image in BGR format, but plt shows image in RGB format. So we converted our color from BGR to RGB using cv2.cvtColor() method.

```python
plt.style.use('seaborn')
```

```python
plt.figure(figsize=(6,4))
plt.grid(False)
plt.imshow(im) # as RGB Format
plt.show()
```

- This is our original image and we have to extract most dominant k colors from it.
- But before that we need to **flatten each channel of image from 2d-->1D**

```python
# Flatten Each channel of the Image
all_pixels  = im.reshape((-1,3))  ## 2nd column is 3 and calculate 1st column automativally
## or we can do im.reshape((330*500,3))
print(all_pixels.shape)
```

- After flattening each channel of image from 2d-->1d, we get array of size: (165000,3) means 3 column vectors where each column is a flattened channel of image.

- Now we will use **k-means clustering** to extract most dominant colors.(See below how it's done)

```python
from sklearn.cluster import KMeans
```

- We will create k clusters 

```python
dominant_colors = 4

km = KMeans(n_clusters=dominant_colors)
km.fit(all_pixels)  ## do training 
```

```python
centers = km.cluster_centers_
centers
```

- These are floating point values bcoz k-means finds centers as mean values. We need to convert these values in integers as pixels only take integer values of colors. And in this case: we are using RGB format, So every channel of each pixel contains value between: (0-255).

```python
import numpy as np
centers = np.array(centers,dtype='uint8')  ## uint8: unsigned integer of 8bits
## we only need to represent color values from 0-255. So 8-bit integer is sufficient
print(centers)
```

- These centers are RGB values of our most dominant 4-colors. The 4th dominant color here is (1 0 1) which means it must be black color or we can say very close to black. And we can see in our original image: this image contains so much black part in lower side.
- Let us see what these 4colors are:

### Plot what all colors are these ?
- Let us plot a (1x4) subplot. And we will show all these 4colors in these 4subplots

```python
plt.figure(0,figsize=(8,2))

i = 1
colors = []
for each_col in centers:
    plt.subplot(1,4,i)  ## this is ith subplot
    plt.axis("off") # turn off the axis values(bcoz it makes our output crowded which look uncool)
    i+=1
    
    # Color Swatch: means a small sample of fabric to demonstrate the look of larger piece
    # We will create a (100x100) small image for all 4dominant colors and fill all pixels of 
    ## (100x100) small sample with ith color(see output). These are color swatches for 4colors.
    
    a = np.zeros((100,100,3),dtype='uint8') ## create 100x100 empty array
    a[:,:,:] = each_col # in ith iteration fill all the pixels of small sample img with ith color
    # In above line: fill all pixels with same RGB values(each_color)
    plt.imshow(a)  ## show this this color swatch in ith subplot
    
plt.show()
```

- These are our 4 most dominant colors in the elephant image.
- Now let us convert our original image into segmented image:

## Segmenting Our Original Image

```python
new_img = np.zeros((330*500,3),dtype='uint8')

print(new_img.shape)
```

```python
centers
```

```python
colors
```

```python
km.labels_
```

```python

for ix in range(new_img.shape[0]):
    new_img[ix] = colors[km.labels_[ix]]
    
new_img = new_img.reshape((original_shape))
plt.imshow(new_img)
plt.show()
    
    
```

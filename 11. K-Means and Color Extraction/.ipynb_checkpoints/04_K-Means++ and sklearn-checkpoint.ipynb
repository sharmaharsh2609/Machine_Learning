### K-Means++

- K-Means is quite sensitive to initalization, if the init is not good, our algorithm is not able to make desired number
of clusters.
- To overcome this problem, we use technique called K-Means++ (described in paper [Robust Seed Selection for K-Means type of Algorithms](https://arxiv.org/abs/1202.1585))
    which chooses initial centers so that they are statiscallly close to final ones.

- The method is default choice for sklearn, so we will use it directly without going into much maths!


Two different types of clustering oberserved for two different types of initialisations as discussed in the last example.

## Using Sci-kit Learn

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y  = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
from sklearn.cluster import KMeans
```

```python
KMeans?
```

- If we see documentation of k-means: press shft+tab, then there is a kwarg: `init='kmeans++'` . This argument takes initial values as selected by kmeans++ . **k-means++ selects initial values in a smarter way**. We can **see descreption of k-means++ on google or sklearn documentation.**
- There are many more arguments for Kmeans which we can see by pressing shft+tab in above cell.

```python
kmeans = KMeans(n_clusters=5)  ## create an instance of KMeans
```

```python
kmeans.fit(X,y)
```

- Now our training part is done. In training we assign random points to all clusters effectively using KMeans++ here.
- We can also see centers assigned to all the clusters by using **obj.cluster\_centers\_** attribute.

```python
centers = kmeans.cluster_centers_
centers
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(5,3))
plt.scatter(X[:,0],X[:,1])
plt.scatter(centers[:,0],centers[:,1],marker='*',color='orange')
plt.show()
```

- Now we can see: initial values of our centers are very much close to final values. It looks like it is final form. This is bcoz sklearn.KMeans use kmeans++ to initialize values which initializes values such that they are very close to the final values.
- We can try to fit data again and recheck these centers again. There will be different values, but they will also be very close to all these centers.

```python
### make predictions(Testing part)
pred = kmeans.labels_
print(pred)
```

- Making predictions assigns some id to each point. Each cluster have their own id from (0 to k-1) and these ids are assignedd to each of the points.
- This implementation is different from the one where we implemented. We added all points belonging to ith cluster in the list of ith cluster. But here, all points are assigned the ids of clusters.

```python
## Let us see points and centers now
plt.figure(figsize=(5,3))
## give color according to the id given to each point as "c=pred"
plt.scatter(X[:,0],X[:,1],c=pred, cmap=plt.cm.Accent)  
plt.scatter(centers[:,0],centers[:,1],marker='*',color='black')
plt.title("Prediction")
plt.show()
```

- Don't worry about colors. Colors are randomly assigned to each cluster everytime bcoz class_id is randomly assigned. **c=pred : pred values are not in exact order of X=[x1, x2,..]. pred are made by picking random points from X array. So there is no fixed order of this. If we fit our data again, then "pred=" will have different oder of points. That's why colors for each clusters will change if we fit data again**. 

- Now these are final values of centers and all the points assigned to each centers.
- Now let us plot our points with actual y_labels(classes) that we created in the beginning:

```python
plt.figure(figsize=(5,3))
plt.scatter(X[:,0],X[:,1],c=y, cmap=plt.cm.Accent)
plt.title("Actual")
plt.show()
```

- Here we are using actual y which will have same order of points and thus same "c=y" always. So it will generate same colors to every cluster always. But it may differ from colors assigned in predictions. Bcoz maybe a point from 3rd cluster is predicted first. Then first color will be given to 3rd cluster instead of 1st cluster. That's why colors assigned to predictions are different from these above.

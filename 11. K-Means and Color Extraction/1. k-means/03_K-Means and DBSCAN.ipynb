## Problem with K-Means
- Can fail sometimes on complex datasets

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles,make_moons
```

```python
X,Y = make_moons(n_samples=200,noise=0.10)
```

```python
## Let us see what this dataset looks like:
plt.style.use('seaborn')
plt.figure(figsize=(5,3))
  ## color according to actual labels for now to visualize data: how it looks like actually
plt.scatter(X[:,0],X[:,1],c=Y, cmap=plt.cm.Accent) 
plt.show()
```

### KMeans on make_moons: 

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)  # create instance of KMeans() class
km.fit(X)   ## do the training
```

```python
print(km.cluster_centers_)
print(km.labels_)  ## as we have only 2classses here, so labels will be only 0 and 1
```

```python
centers = km.cluster_centers_
label = km.labels_

plt.figure(figsize=(5,3))
plt.scatter(X[:,0],X[:,1],c=label, cmap=plt.cm.Accent)  # give color according to the label
plt.scatter(centers[:,0],centers[:,1],color='red', marker='*')
plt.show()
```

## Dominant Color Extraction for Image Segmentation
- **Image Segmentation**
- Segmentation partitions an image into regions having similar visual appearance corresponding to parts of objects
- We will try to extract the most dominant 'K' Colors using K-Means
- We can apply K-Means with each pixel will reassigned to the closest of the K Colors, leading to segmentation

Fig : Bishop, Pattern Recognition and ML

### Steps for Image segmentation:
Reading Image ⇒ Reshaping into a 2D Array ⇒ Applying K-Means ⇒ Extracting Dominant Colors ⇒ Assigning these colors to the new image ⇒ Reshaping back to 3D

```python
import matplotlib.pyplot as plt
import cv2
```

```python
im =  cv2.imread('elephant.jpg') #Reads an image into BGR Format

im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
print(im.shape)
```

```python
plt.style.use('seaborn')
```

```python
plt.figure(figsize=(6,4))
plt.grid(False)
plt.imshow(im) # as RGB Format
plt.show()
```

```python
# Flatten Each channel of the Image
all_pixels  = im.reshape((-1,3))  ## 2nd column is 3 and calculate 1st column automativally
## or we can do im.reshape((330*500,3))
print(all_pixels.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
dominant_colors = 4

km = KMeans(n_clusters=dominant_colors)
km.fit(all_pixels)  ## do training 
```

```python
centers = km.cluster_centers_
colors = centers
colors
```

```python
import numpy as np
colors = np.array(colors,dtype='uint8')  ## uint8: unsigned integer of 8bits
## we only need to represent color values from 0-255. So 8-bit integer is sufficient
print(colors)
```

### Plot what all colors are these ?
- Let us plot a (1x4) subplot. And we will show all these 4colors in these 4subplots

```python
plt.figure(0,figsize=(8,2))

i = 1  ## for each subplot, increment it in each iteration
for each_col in colors:
    plt.subplot(1,4,i)  ## this is ith subplot
    plt.axis("off") # turn off the axis values(bcoz it makes our output crowded which look uncool)
    i+=1
    
    # Color Swatch: means a small sample of fabric to demonstrate the look of larger piece
    # We will create a (100x100) small image for all 4dominant colors and fill all pixels of 
    ## (100x100) small sample with ith color(see output). These are color swatches for 4colors.
    
    a = np.zeros((100,100,3),dtype='uint8') ## create 100x100 empty array
    a[:,:,:] = each_col # in ith iteration fill all the pixels of small sample img with ith color
    # In above line: fill all pixels with same RGB values(each_color)
    plt.imshow(a)  ## show this this color swatch in ith subplot
    
plt.show()
```

## Segmenting Our Original Image

```python
new_img = np.zeros((330*500,3),dtype='uint8')  ## create emtply img of same size as original img
print(new_img.shape)
```

```python
colors  ## colors stores our most dominant 4colors
```

```python
print(km.labels_)
print(km.labels_.shape)
print(np.unique(km.labels_))
```

```python
plt.figure(figsize=(6,4))
for ix in range(new_img.shape[0]):  ## for every pixel 
    new_img[ix] = colors[km.labels_[ix]] # value of pixel = nearest dominant color(given by label)
plt.grid(False)
original_shape = im.shape ## original image shape
segmented_img = new_img.reshape((original_shape))
plt.imshow(segmented_img)
plt.show()
```

```python
elephants = np.full(im.shape, 255)  ## create empty image (all white: 255 value)
labels = km.labels_.reshape(im.shape[0:2])  ## convert from 165000 to 330,500 array

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        if labels[i,j] == 2:
            elephants[i,j,:] = im[i,j,:]

plt.figure(figsize=(6,4))
plt.grid(False)
plt.imshow(elephants)
plt.show()
```

```python
labels = km.labels_.reshape(im.shape[0:2])  ## convert from 165000 to 330,500 array
def plot_func(label_value):
    extracted_img = np.full(im.shape, 255)  ## create empty image (all white: 255 value)
    
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            if labels[i,j] == label_value:
                extracted_img[i,j,:] = im[i,j,:]

    plt.grid(False)
    plt.imshow(extracted_img)

    
## draw a subplot with all 4 colotrs
plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plot_func(i)  ## pass i as label_value
plt.show()
print(colors)
```

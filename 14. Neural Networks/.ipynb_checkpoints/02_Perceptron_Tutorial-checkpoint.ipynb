# Perceptron 
#### (Artificial Neuron)

- A Perceptron is a single layer neural network
- It acts a Linear Classifier, Binary classification
- Loss Function : Binary Cross Entropy   
(-ve of Log Likelihood)
- Optimisation : Gradient Descent/ Stochastic Gradient Descent

## Model and helper functions

```python
def sigmoid(z):
    return (1.0)/(1+np.exp(-z))
```

```python
a = np.array([-200,10,0,10,200])
sigmoid(a)
```

## Implement Perceptron Learning Algorithm
- Learn the weights
- Reduce the loss 
- Make the Preditctions

```python
def predict(X,weights):
    """X -> (m,(n+1)) matrix, w -->((n+1),1) vector"""
    # so result of X.w = (m,1). So it multiplies 1st row of X with all w vector, then x2 with w and so on..
    # so X.w = x0.w0+x1w1+...xnwn for 1st example, then in second iteration it is with 2nd and so on..
    ## and finally result is (m,1)=> m results: we find w.T*x+b for m examples seperately
    z = np.dot(X,weights)   ## do this all in a single step using vectorization instead of for loop
    predictions = sigmoid(z)  ## predict results for whole array in a single go instead of for loop
    return predictions

def loss(X,Y,weights):
    """Binary Cross Entropy"""
    ## -1/m * summation[yi.log(yi) + (1-yi).log(1-yi)]
    # so we find -yi.log(yi)-(1-yi).log(1-yi) and then do mean which is same as 1/m* sum(..)
    Y_ = predict(X,weights)
    cost = np.mean(-Y*np.log(Y_) - (1-Y)*np.log(1-Y_))
    return cost
## this was vectorized implementation without any for loop.

def update(X,Y,weights,learning_rate):
    """Perform weight updates for 1 epoch"""
    Y_ = predict(X,weights)
    dw = np.dot(X.T,Y - Y_) ## see with copy-pen(draw matrices) what we are doing here.
    ## dw = summation{ (y_[i]-y[i]) * x[j][i]} => for every jth feature we sum over all ecamples
    ## and grad = -1/m * summation{...} = -1/m * dw = -dw/m
    # so weights = weights - (lr * -dw/m)
    
    m = X.shape[0]
    weights = weights + learning_rate*dw/(float(m))
    return weights

def train(X,Y,learning_rate=0.5,maxEpochs=1000):
    #Modify the input to handle the bias term
    ones = np.ones((X.shape[0],1))
    X = np.hstack((ones,X))
    #Init Weights 0
    weights = np.zeros(X.shape[1]) #n+1 entires
    
    #Iterate over all epochs and make updates
    for epoch in range(maxEpochs):
        
        weights = update(X,Y,weights,learning_rate) # call this func to upate theta in 1 epoch
        
        if epoch%10==0: # print loss after every 10epochs
            l = loss(X,Y,weights)
            print("Epoch %d Loss %.4f"%(epoch,l))
    return weights    
```

- Try with different learning rates for training. So using high learning rate will reach results faster.

```python
weights = train(X,Y,learning_rate=0.8,maxEpochs=1000)
```

 
### Perceptron Implementation - Part-II 
- Make Predictions
- Visualise Decison Surface
- Linear vs Non-linear classification

```python
def getPredictions(X_Test,weights,labels=True):
    if X_Test.shape[1] != weights.shape[0]:  ## check if X_test has n features or (n+1)
        ones = np.ones((X_Test.shape[0],1)) ## if not n+1, then add a columns of 1s in beginning 
        X_Test = np.hstack((ones,X_Test))

    probs = predict(X_Test,weights)
    
    if not labels:  ## if labels is passed False in arg of function, then return on probs
        return probs
    else:           ## else return on;y labels which we can calculate as below: 
        labels = np.zeros(probs.shape)
        #create array of 0s and replace values with 1 at indices where prob>0.5 
        labels[probs>=0.5] = 1  ## this is again vectorization kind of thing instead of for loop
        return labels
```

```python
x1 = np.linspace(-12,2,10) ## generate 10 data points with same distance in range -12 to 2
x2 = -(weights[0] + weights[1]*x1)/weights[2]
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Accent)
plt.plot(x1,x2,c='red')
plt.show()
```

```python
# Find the accuracy!
Y_ = getPredictions(X,weights,labels=True)
training_acc = np.sum(Y_==Y)/Y.shape[0]
print(training_acc)
```

- Accuracy is 100%

#### Now let us try this single layer Neural Network with sigmoid as activation function on make_moons(which is linearly unseperable data)

```python
from sklearn.datasets import make_moons
```

```python
X,Y = make_moons(n_samples=1000, shuffle=True, noise=0.2, random_state=1)
print(X.shape, Y.shape)
```

```python
plt.figure(figsize=(5,3))
plt.style.use("seaborn")
plt.scatter(X[:,0],X[:,1],c=Y,cmap= plt.cm.Accent)
plt.show()
```

```python
weights = train(X,Y,learning_rate=0.8,maxEpochs=1000)
```

```python
x1 = np.linspace(-2,3,10) ## generate 10 data points with same distance in range -2 to 3
x2 = -(weights[0] + weights[1]*x1)/weights[2]
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Accent)
plt.plot(x1,x2,c='red')
plt.show()
```

```python
# Find the accuracy!
Y_ = getPredictions(X,weights,labels=True)
training_acc = np.sum(Y_==Y)/Y.shape[0]
print(training_acc)
```

- Now accuracy is 87% because this dataset is not linearly seperable.

### Perceptron can't seperate these kind of datasets!!

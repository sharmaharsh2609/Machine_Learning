```python
import numpy as np
```

```python
#Model Parameters
input_size = 3 # no_of_input_features in input layer
layers = [4,3] # no of neurons in 1st and 2nd layer(hidden layers)
output_size = 2
```

```python
def softmax(a):
    e_pa = np.exp(a) # np.exp(a) applies exponential on each term of array. this is e_power(a) array.
    ans = e_pa/np.sum(e_pa,axis=1,keepdims=True)
    return ans    
```

```python
class NeuralNetwork:
    
    def __init__(self,input_size,hidden_layers,output_size):
        np.random.seed(0)
        
        model = {} #Dictionary to store model parameters
        
        #First Layer (initialize weights)
        model['W1'] = np.random.randn(input_size,hidden_layers[0])
        model['b1'] = np.zeros((1,hidden_layers[0]))
        
        #Second Layer
        model['W2'] = np.random.randn(hidden_layers[0],hidden_layers[1])
        model['b2'] = np.zeros((1,hidden_layers[1]))
        
        #Third/Output Layer
        model['W3'] = np.random.randn(hidden_layers[1],output_size)
        model['b3'] = np.zeros((1,output_size))
        
        self.model = model
        self.activation_outputs = None  
    
    def forward(self,x):
        
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']  ## unpack w1,w2,w3 from 'model'dictionay.
        b1, b2, b3 = self.model['b1'],self.model['b2'],self.model['b3']
        
        #  for H1(1st hidden layer): input = X
        z1 = np.dot(x,W1) + b1  
        a1 = np.tanh(z1)  
        
        z2 = np.dot(a1,W2) + b2  ## Z2(array for weighted sum for all examples)
        a2 = np.tanh(z2)
        
        z3 = np.dot(a2,W3) + b3  ## Z3(weighted sum for 3rd layer)
        y_ = softmax(z3)   ## for final layer apply softmax() instead of activation func. tanh()
        
        self.activation_outputs = (a1,a2,y_) ## store activation output of all n-layers in form of
        # a tuple (a1,a2...an) 
        
        return y_
        
    def backward(self,x,y,learning_rate=0.001):
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3'] # unpack model parameters
        b1, b2, b3 = self.model['b1'],self.model['b2'],self.model['b3']
        m = x.shape[0] ## no. of examples in x
        
        a1,a2,y_ = self.activation_outputs # get values of ouputs of each layer from "self.acti.."
        
        delta3 = y_ - y  
        
        dw3 = np.dot(a2.T,delta3)
        db3 = np.sum(delta3,axis=0)/float(m)
        
        delta2 = np.dot(delta3,W3.T)*(1-np.square(a2))  ## * means element wise multiplication
        dw2 = np.dot(a1.T,delta2) # d(Loss)/dw[l] = a[l-1]^T.delta[l]
        db2 = np.sum(delta2,axis=0)/float(m) # d(Loss)/db[l] = 1/m*[np.sum(delta[l])]
        
        delta1 = (1-np.square(a1))*np.dot(delta2,W2.T)
        dw1 = np.dot(X.T,delta1)
        db1 = np.sum(delta1,axis=0)/float(m)
        
        #Update the Model Parameters using Gradient Descent
        self.model["W1"]  -= learning_rate*dw1
        self.model['b1']  -= learning_rate*db1
        
        self.model["W2"]  -= learning_rate*dw2
        self.model['b2']  -= learning_rate*db2
        
        self.model["W3"]  -= learning_rate*dw3
        self.model['b3']  -= learning_rate*db3

    def predict(self,x):
        y_out = self.forward(x)
        return np.argmax(y_out,axis=1)
    
    def summary(self):  # to show summary after each step
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']
        a1,a2,y_ = self.activation_outputs
        
        print("W1 ",W1.shape)
        print("A1 ",a1.shape)
        
        print("W2 ",W2.shape)
        print("A2 ",a2.shape)
        
        print("W3 ",W3.shape)
        print("Y_ ",y_.shape)    
```

```python
def loss(y_oht,p):
    l = -np.mean(y_oht*np.log(p)) # categorical entropy loss.
    return l

def one_hot(y,depth): # depth is number of classes.  
    m = y.shape[0]
    y_oht = np.zeros((m,depth))
    y_oht[np.arange(m),y] = 1
    # np.arrange will give index for all rows from 0,1,2..m and y will give index for class.
    return y_oht
```

## Training Our Model
- In training, we make a forward pass, then we see the final loss and then we make a backward pass which calculates loss for each layer and modify 'w' parameters to decrease that error using gradient descent algo. And we do these steps in every epoch repeatedly until the loss is not minimized.

```python
model  = NeuralNetwork(input_size=2,hidden_layers=[10,5],output_size=2)
```

```python
model.forward(X) # we need at least 1 forward pass to see summary.
model.summary()  ## see summary after one forward pass
```

```python
def train(X,Y,model,epochs,learning_rate,logs=True):
    training_loss = []  # to store loss in each step 
    
    classes = 2  # there are only 2 classes in our data
    Y_OHT = one_hot(Y,classes) # convert Y to one hot matrix
    
    for ix in range(epochs):
        
        Y_ = model.forward(X) # perform forward pass
        l = loss(Y_OHT,Y_)  # see final loss after each forward pass
        training_loss.append(l) # add final loss after each step in a list to visualize ut later 
        # see loss for all layers and update 'w' using gradient descent
        model.backward(X,Y_OHT,learning_rate)
        
        if(logs):  # if logs is passed true in argument, then print loss after each step else not
            print("Epoch %d Loss %.4f"%(ix,l)) # 0.4f means upto 4digits after decimal
    
    return training_loss
```

```python
losses = train(X,Y,model,1000,0.001)
```

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
from sklearn.datasets import make_moons,make_circles,make_classification
```

```python
def load_dataset(dataset):
    if dataset=='moons':
        X,Y = make_moons(n_samples=500,noise=0.2,random_state=1) #Perceptron
    elif dataset=='circles':
        X,Y = make_circles(n_samples=500, shuffle=True, noise=0.2, random_state=1, factor=0.2)
    elif dataset=='classification':
        X,Y = make_classification(n_samples=500,n_classes=2,n_features=2,n_informative=2,n_redundant=0,random_state=1)
    else:
        #Create XOR Dataset
        X = np.array([[0,0],
                    [0,1],
                    [1,0],
                    [1,1]])
        Y = np.array([0,1,1,0])
        
    return X,Y
```

```python
datasets = ["xor","classification","moons","circles"]

for d in datasets:
    model = NeuralNetwork(input_size=2,hidden_layers=[8,4],output_size=2)
    X,Y  = load_dataset(d) # get data from load_dataset() function we defined earlier
    train(X,Y,model,1000,0.001,logs=False) # do training and logs=False to not print logs
    outputs = model.predict(X)
    
    training_accuracy = np.mean(outputs==Y)
    print("Training Acc %.4f"%training_accuracy)
    
    plt.title("Dataset "+d)
    plot_decision_boundary(lambda x:model.predict(x),X,Y)
    plt.show()
        
```

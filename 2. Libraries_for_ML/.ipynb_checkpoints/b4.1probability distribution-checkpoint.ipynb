#### Probability Distribution:
- A probability distribution is a statistical function that describes all the possible values and likelihoods that a random variable can take within a given range.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xsd = np.random.randn(100)  # sample from standard normal distribution
print(xsd)
```

```python
# generate a data of 250 students' physics marks with mean 70 and std=5
Xsnd = np.random.randn(250) #generates 50 points from standard normal distribution. 

u = 70     
sigma = 5 
physics = Xsnd*sigma + u  # change standard normal distribution to desired normal distribution.   
print(physics)
```

```python
plt.style.use('seaborn')
plt.hist(physics, bins=10)
plt.show()
```

```python
Xsnd = np.random.randn(150) 

u = 70     
sigma = 1 
physics = Xsnd*sigma + u  # change standard normal distribution to desired normal distribution.   
plt.hist(physics, bins=10)
plt.show()
```

- See x-axis now. It has spread around 69-71. Now spread is decreased.

```python
physics = np.round(physics)  # we round off bcoz unique() will give all values unique otherwise.
z = np.unique(physics, return_counts=True)
print(z)
```

- This was a univariate normral distribution.

```python
# let us draw a univariate distribution with y-axis as zeros
Xsnd = np.random.randn(150) 
u = 70     
sigma = 5 

x = Xsnd*sigma + u  
y = np.zeros(150)

plt.scatter(x, y)
plt.show()
```

```python
#Bivariate
mean = np.array([0.0, 0.0]) # centered around origin
cov = np.array([[1,0], [0,1]])  #covariance matrix

dist = np.random.multivariate_normal(mean, cov, 500)
dist
```

```python
plt.scatter(dist[:,0], dist[:,1])
plt.show()
```

```python
mean = np.array([0.0, 0.0]) # centered around origin
cov = np.array([[1,0], [0,1]])  #covariance matrix
dist = np.random.multivariate_normal(mean, cov, 500)

mean2 = np.array([6.0, 5.0]) # centered around origin
cov2 = np.array([[1.3,0.7], [0.7,1.1]])  #covariance matrix
dist2 = np.random.multivariate_normal(mean2, cov2, 500)

plt.scatter(dist[:,0], dist[:,1])
plt.scatter(dist2[:,0], dist2[:,1])
plt.show()
```

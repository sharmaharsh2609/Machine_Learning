## Generating Datasets

Some useful functions to generate datasets available in `sklearn.datasets`

- `make_regression`
- `make_classification`
- `make_blobs`
- `make_moons`


```python
from sklearn.datasets import make_regression,make_classification,make_blobs,make_moons
```

```python
from matplotlib import pyplot as plt
```

```python
X,Y = make_regression?
```

- It is simply a function which returns us the data from a randomly generated regression problem.

```python
X,Y = make_regression()
X.shape, Y.shape
```

```python
X,Y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=10.8)
print(X.shape,Y.shape)
```

```python
plt.style.use('seaborn')
plt.scatter(X,Y)
plt.show()
```

```python
make_classification?
```

- It generates a random n-class classification problem

```python
XC,YC = make_classification(n_samples=400,n_features=2,n_classes=3,n_clusters_per_class=1,n_informative=2,n_redundant=0,random_state=5)
```

```python
print(XC.shape,YC.shape)
```

- Thus we have 400 examples and 2 features

```python
import numpy as np
np.unique(YC)
```

- Thus we have 3 classes

```python
plt.scatter(XC[:,0],XC[:,1],c=YC, cmap="brg")
plt.show()
```

### Blobs Dataset

```python
XB,YB = make_blobs?
```

- Generate isotropic Gaussian blobs for clustering.

```python
XB,YB = make_blobs()
XB.shape, YB.shape
```

```python
XB,YB = make_blobs(n_samples=1000, n_features=2, centers=5,random_state=10)
XB.shape, YB.shape
```

```python
plt.scatter(XB[:,0],XB[:,1],c=YB, cmap="brg")
plt.show()
```

### Moons Dataset

```python
XM, YM = make_moons?
```

- Make 2 interactive half circles

```python
XM, YM = make_moons()
XM.shape, YM.shape
```

```python
XM,YM = make_moons(n_samples=400,noise=0.2, random_state=5)
XM.shape, YM.shape
```

```python
plt.scatter(XM[:,0],XM[:,1],c=YM, cmap="rainbow")
plt.show()
```

## Subplots

```python
plt.figure(figsize=(8,8))

plt.subplot(221)
plt.title("Regression")
plt.scatter(X,Y, cmap="brg")

plt.subplot(222)
plt.title("Classification")
plt.scatter(XC[:,0],XC[:,1],c=YC, cmap="brg")



plt.subplot(223)
plt.title("Blobs")
plt.scatter(XB[:,0],XB[:,1],c=YB, cmap="brg")


plt.subplot(224)
plt.title("Moons")
plt.scatter(XM[:,0],XM[:,1],c=YM, cmap="brg")

plt.show()
```

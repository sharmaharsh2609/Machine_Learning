### Distance formula:
- We need some functions in our knn implementation. One of them is distance formula. LEt us see this.

```python
def distance(p1, p2):
    return ( (p1[0]**2-p2[0])**2 + (p1[1]-p2[1])**2 )**0.5
```

```python
print(distance(X_data[0], X_data[1]))
print(distance(X_data[0], X_data[2]))
```

- We can see the distance of 1st point from 2nd is larger than distance of 1st point from 3rd.
- **We can find distance of all the data points from our test point and then sort those distances and then pick the top k-points**

- We can either write function or class for prediction. But classes are better as we will use classes in 'sklearn' also.

#### Explanation of below code:
- k is passed by user
- **There is no training involved in knn algo. So fit() method simply stores values of X_data and Y_data.(Read this statement after completing all the algorithm in future. After studying all algos, come here again and read this all. fit() method do not make any sense for now).
- **predict()** method is most imp. method and core of the program:
- Here we create empty list of distances. Then we find distances of all the points in the dataset from the test point and store them inside distances. But we need their track also. So, we create a list of tuples where we append (distance, index) for every point
- Then sort the distances and store sorted distance and index as a **list of tuples**.
- **distances[:,self.k]** gives us top k points whose distance is shorttest as array is sorted.
- But we don't need top closest points or top k shortest distances. **We only need classes of these points. From which class do they all belong. And then do voting and find probability**.
- We iterate over the list of tuples, in which we have top k points. We store class of each of those top k points inside list named 'classes'. These labels are given in label vector 'Y_data'.
- **np.unique()** gives us the counts for all the classes present in list 'classes' and their counts respectively.(Here we only have 2 classes: '0.0' and '1.0')
- **np.argmax(classes)**  gives us index of class from 'classes' list, **which class has highest number of counts in top k points**.
- We return a dictionary which contains class name from which our test data belongs and probability that it belongs from that class.(Obviously for 2points: prob will be more than 0.5)

```python
class kNNClassifier:
    def __init__(self, k=5):      # default k=5 if user do not pass no. of k points
        self.k = k            # initializing parameter is k in this class.
        
    def fit(self, X, Y):
        self.X, self.Y = X, Y
        
    def distance_formula(self, p1, p2):
        return ( (p1[0]**2-p2[0])**2 + (p1[1]-p2[1])**2 )**0.5
    
    def predict(self, p):  # p is test point
        distances = []  # list of tuples
        for i in range(len(self.X)): #find distance of all points from test point('p')
            distances.append((self.distance_formula(self.X[i], p), i))
            #(distance_formula(self.X[i], p), i): this is a tuple that we append in the list
            
            distances = sorted(distances, key=lambda x: x[0] ) # pass 1st parameter as key.
            # by default, sorted() method sorts acc. to 1st parameter, then 2nd parameter so on..
            top_k_points = distances[:self.k]
            
            classes = []   # create empty list to store all classes in label(which is Y_data)
            for dist, i in top_k_points:
                classes.append(self.Y[i])  #append lable for ith index( for top k points)
                class_ , count = np.unique(classes, return_counts=True)  
            #returns names of classes and corresponding counts by passing 'return_count' argument
                ix = np.argmax(count) #index of argument with maximum counts
                
                pred= {"class": class_[ix], "prob": count[ix]/np.sum(count)} # a dictionary
                return pred
#returns dict which has class name and probability for the test point to belong from that class.
```

```python
knn_model = kNNClassifier(7)
knn_model.fit(X_data, Y_data)
knn_model.predict(X_data[0])
```

- Here probability for the point X_data is '1.0'. This is because all the 7 closest points to the testing data belong to the same class '0.0'
- X_data[0] and Y_data[0] is training data and not testing data. But here we used it for understanding only.

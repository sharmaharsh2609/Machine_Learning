## Recommendation System:


### Classification:
- We generally devide our recommendation systems into 2types:
    - Content Based R.S.
    - Collaborative filtering R.S.

### Content Based Recommendation System:
- It is based on the similarity between items calculated using the rating users have given to items.
<img src="07.2item_based.png" width=400>
- eg: Here both user1 and user2 buy orange and an apple. Now a 3rd user come to shop and buy apple. So orange is also recommended to user3 as all the users who bought apple also bought orange(means apple and orange items are very similar). 
- Here we are recommending based on the items. We are checking similarities between items. Thats why it is called item-based R.S.

### Collaborative Filtering:
- It filters data based on the **behaviour of user rather than on items-based**.
<img src="07.3collaborative.png" width=400>
- It is again divided in 2parts:
    - Model Based
    - Memory Based

### Build Recommendation System:


```python
movies = df.pivot_table(index="user_id", columns="title", values="rating")
# it reshapes dataframe
movies
```

```python
# we can see movies with most no of ratings from 'ratings' dataframe:
ratings['no of ratings'].sort_values(ascending=False)
```

- Here we can see star_wars movie has highest number of ratings.

#### Let us recommend movies to a person who watches and likes starwar movie

```python
# from 'movies' dataframe, we can see ratings given by all the users to 'Starwars' movie:
starwar_ratings = movies['Star Wars (1977)']
starwar_ratings  # It is a pandas series which as we passed one column of the dataframe
# check by type(starwar_ratings)
```

```python
# now we can find the correlation of this 'starear_ratings' with whole ratings of all the movies:
movies.corrwith(starwar_ratings)  #see explanation below first
```

```python
# we can store it in a dataframe. Let's convert above series into dataframe
corr_starwars = pd.DataFrame(movies.corrwith(starwar_ratings), columns=["Correlation"])
corr_starwars
```

```python
corr_starwars.dropna(inplace=True)
corr_starwars
```

```python
corr_starwars.sort_values(by="Correlation", ascending=False)
```

```python
# let us join corr_starwars with ratings dataframe:
corr_starwars = corr_starwars.join(ratings['no of ratings'])  # This method join col(s) of other dataframes
corr_starwars
```

```python
corr_starwars[corr_starwars['no of ratings']>100].sort_values(by='Correlation', ascending=False)
# "corr_starwars['no of ratings']>100": This command returns series of True and Fasle and then 
# corr_statement[..]: will print all those values which return True

#It prints all the values for which 'no of rating'>100. 
```

- Let us create our generalized final Prediction function from all above data.

## Prediction function


```python
def predict_movie(movie_name):
    movie_ratings = movies[movie_name] # create a series of user ratings for 'movie_name'
    #create correlation of df for movie_ratings with all the movies and pass the col name='correlation'
    corr_movie = pd.DataFrame(movies.corrwith(movie_ratings), columns=['correlation'])
    corr_movie.dropna(inplace=True)   #drop all the NaN values.
    
    corr_movie = corr_movie.join(ratings['no of ratings'])  # join no of rating col also to corr_movie
    #remove less than 100 rating movies and sort in descending order
    prediction = corr_movie[corr_movie['no of ratings']>100].sort_values(by='correlation', ascending=False)
    
    return prediction #it stores all the movie names with more than 100ratings and high correlation
```

```python
predictions = predict_movie("Titanic (1997)")
predictions.head()  #print top 5 similar movies
```

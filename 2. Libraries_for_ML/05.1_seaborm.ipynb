## Seaborn
- Seaborn is a library used for data visualization with the help of numpy, matplotlib, pandas.

`pip install seaborn`

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')  # this seaborn is a theme of matplotlib.pyplot. It is totally 
# different from the seaborn library. They just have same names. That's all. 
```

- To plot something, we first need some data.
##### - sns.load_dataset()
- This method can load any dataset from a online repository for testing. There is a github repo: https://github.com/mwaskom/seaborn-data (it is given in the documentation) . 

```python
tips_data = sns.load_dataset('tips')   #extension is not required
print(tips_data)
```

### 1. BAR PLOT


```python
plt.figure(figsize=(5,5))
sns.barplot(x='sex', y='tip', data=tips_data)
plt.show()
```

```python
#check mean for male and female
# this below operation can be done in pandas dataset.
# print(tips_data[tips_data['sex']=='Female']) #print all the data with sex="female"
# print(tips_data[tips_data['sex']=='Female']['tip']) #print tip for all female
print(tips_data[tips_data['sex']=='Female']['tip'].mean())  #mean for female
print(tips_data[tips_data['sex']=='Male']['tip'].mean())    #mean for male
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='sex', y='tip', data=tips_data, estimator=np.std) #assign std func to estimator
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='sex', y='tip', hue='time', data=tips_data, estimator=np.std)
plt.show()
```

### 2. Count Plot:

```python
plt.figure(figsize=(5,5))
sns.countplot(x='sex', data=tips_data)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(y='sex', data=tips_data)
plt.show()
```

```python
# another example with hue
plt.figure(figsize=(5,5))
sns.countplot(x='day', hue='sex', data=tips_data)
plt.show()
```

### 3. Box Plot
- Box plot is one of most used in Data Visualization and very important.

```python
sns.boxplot(x='sex', y='tip', data=tips_data)
plt.show()
```

- Now let us try it with hue:

```python
sns.boxplot(x='day',y='tip', hue='sex', data=tips_data)
plt.show()
```

### 4. Violin Plot


```python
sns.violinplot(x='day', y='tip', hue='sex', data=tips_data)
plt.show()
```

```python
sns.violinplot(x='day', y='tip', hue='sex', data=tips_data, split=True)
plt.show()
```

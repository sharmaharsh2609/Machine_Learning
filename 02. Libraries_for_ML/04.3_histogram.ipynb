#### np.random.randn()


```python
xsd = np.random.randn(100)  # sample from standard normal distribution
print(xsd)
```

```python
x = np.random.randn(10000000)
```

```python
print(x.mean(), x.std())
```

```python
Xsnd = np.random.randn(100) # sample of 100elements from standard normal distribution.

u = 60     #customized mean
sigma = 5  #customized standard deviation that we want.
maths = np.round(Xsnd*sigma + u)   #np.round rounds off all the float values
print(maths)
```

```python
plt.style.use("seaborn") #let us use seaborn to plot histogram.

plt.hist(maths)
plt.xlabel("Marks range:")
plt.ylabel("Prob/freqCount of students")
plt.title("Histogram")
plt.show()
```

```python
physics = np.round(Xsnd*5+40)  # let us have mean of physics =40
maths = np.round(Xsnd*5+60)

plt.hist(physics)
plt.hist(maths, alpha=0.7)
plt.xlabel("Marks range:")
plt.ylabel("Prob/freqCount of students")
plt.title("Histogram")
plt.show()
```

## Bar Graphs

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn')   #let us use seaborn theme for all our programs
```

```python
plt.figure(figsize=(5,5))

x = [0,2,4]
h = [10,20,15]
plt.bar(x, h)
plt.show()
```

```python
plt.figure(figsize=(5,5))

x = np.array([0,2,4]) # I chose it as numpyarray to perform x+0.25 operation later (which is not
# allowed in python list). It is also always better to use numpy array. 

plt.bar(x-0.25, [10,20,15], width=0.5)  #we can define width of bars
plt.bar(x+0.25, [20,10,12], width=0.5)
plt.show()
```

##### width parameter:
It defines the width of the bar.eg- above width=0.5 (Default width=0.8)
##### - label argument and legend() method:
We can pass label argument to give name to bar1 and bar2. But we need to use **legend()** method to show it on graph.
##### - tick_label argument:
It is an array which gives names to all the elements of the bar.(See below implementation)
##### - plt.title():
To give title to the graph
##### - plt.xlabel() and plt.ylabel():
To give labels to x-axis and y-axis
##### - plt.xlim() and plt.ylim()
It put some limit to the length of graph to be shown. eg- In above example, we can only see x-axis upto -1 to 4 and y-axis upto 0-20. It shows only upto the last point of x and heigth vectors.<br> To show more part of graph, we can use these 2methods. (See implementation)

```python
plt.bar(x-0.25, [10,20,15], width=0.5, label="Current Year", tick_label=['Gold','Platinum','Silver'])  
plt.bar(x+0.25, [20,10,12], width=0.5, label="Next Year") 
plt.legend()
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.xlim(-2,5)
plt.ylim(0,40)

plt.show()
```

```python
plt.style.use('dark_background')
plt.bar(x-0.25, [10,20,15], width=0.5, color='green', label="Current Year", tick_label=['Gold','Platinum','Silver'])  
plt.bar(x+0.25, [20,10,12], width=0.5, label="Next Year", color='orange') 
plt.legend()
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.show()
```

## Pie Charts


```python
# let us create pie chart for certain subjects. 
subjects = "English", "Maths", "Chemistry", "Physics"   #without any braces is a tuple in python
weightage = [20, 10, 15, 5]  #these are weightage of each subject respectively

plt.pie(weightage, labels=subjects) # weightage is x array and subject is passed as label.
plt.show()
```

Some main parameters are:
##### - label:
It takes a array which contains label names of each wedges of the array.
##### - explode:
It is used to highlight some portion of the array. It takes array with som evalues as argument for each wedge. These values are the distance of that portion from the center.

```python
#example
plt.figure(figsize=(3,3))
plt.pie(weightage, labels=subjects, explode=(0,0,1,0))
plt.show()
```

```python
plt.figure(figsize=(3,3))
plt.pie(weightage, labels=subjects, explode=(0.1,0.1,0.1,0.1)) #highlight all parts
plt.show()
```

##### - shadow
It creates a shadow like structure(See implementation) if we pass True as argument. By default it has value False
##### - color
It takes array-like as argument which contains all the colors.(This is easy. So I've not implemented it below.)
##### - autopct
It takes some string formatting as parameter and represents percentage as output for each label. <br>eg- **autopct=.2f%** means show 2digits after decimals.
##### - plt.title()
It is a plt method used to give title to pie chart

```python
plt.pie(weightage, labels=subjects, explode=(0,0,0.1,0.1), shadow=True, autopct="%.2f%%")
plt.title("Subjects")
plt.show()
```

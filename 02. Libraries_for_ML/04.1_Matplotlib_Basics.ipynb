### plt.figure()
- It is used to define the size of the plot. We can define size of plot by passing **figsize** parameter to plt.figure() 

```python
plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(3,3))
arr = np.arange(4)**2
plt.plot(arr)
plt.show()
# It will take arr as y-axis and [0,1,2,3] as x-axis by default
```

```python
#if we increase value the number of points in array, then it will look like much better.
x = np.arange(10)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# eg
x = np.arange(10)  # this is x array
y1 = x**2         # this is y1 array to plot graph of y1=x^2 
y2 = 2*x + 3      # this is y2 array which plots another graph of y2=2x+3

print("x: {}\ny1: {}\ny2: {}".format(x,y1,y2))
```

```python
plt.figure(figsize=(5,5))
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='green')
plt.show()
```

```python
plt.figure(figsize=(3,3))
plt.plot(x,y1,color='red')
plt.show()
plt.figure(figsize=(3,3))
plt.plot(x,y2,color='green')
plt.show()
```

```python
plt.plot(x,y1,color='red', label="Apple")
plt.plot(x,y2,color='green', label="Mango")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price of fruits over time")
plt.show()
```

```python
plt.plot(x,y1,color='red', label="Apple", marker='o')
plt.plot(x,y2,color='green', label="Mango", linestyle='dashed', marker='*')
plt.legend()
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price of fruits over time")
plt.show()
```

## Scatter Plots
- Scatter plot, plots points on the graph instead of plotting lines.
- It shows discrete points instead of a line

```python
plt.scatter(x,y1)
plt.show()
```

```python
plt.scatter(x,y1,color='red', label="Apple", marker='^')
plt.scatter(x,y2,color='green', label="Mango", linestyle='dashed', marker='*')
plt.legend()
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot: Price of fruits")
plt.show()
```

```python
# another example of a scatter plot of any random 1000 numbers in 1000x1 array.
x = np.random.random((1000,2))

for i in range(x.shape[0]):
    plt.scatter(x[i,0],x[i,1],c='green',marker='+')  # x[0] as x-axis and x[1] as y-axis
    
plt.show()
```

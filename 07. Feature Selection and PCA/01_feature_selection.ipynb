### Why to use feature selection?

→ It enables the machine learning algorithm to train faster, reduces the complexity and makes it easier to interpret.

→ It improves the accuracy of a model if the right subset is chosen.

→ It reduces overfitting.

→ There is a concept of `curse of dimentionality`. It means accuracy increases as we increase number of features. But after a threshhold value our accuracy keeps on decreasing if we increase features any furthur


## Univariate Feature Selection:

```python
import numpy as np, pandas as pd
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

- **chi2** is used for performing CHI square method and **SelectKBest** is used to select top k best features, where k=10 by default. k is a keyword argument

```python
# Read X_train
data = pd.read_csv("./mobile_price_train.csv")
data.head()
```

```python
## seperate X and Y from the data
X_train = data.iloc[:,:-1]  # all columns upto last one(last excluded)
y_train = data.iloc[:,-1]   # output is pandas.series not dataframe as it is 1D(see shape of y)
print(X_train.shape, y_train.shape)
```

```python
## print y_train by converting it to Dataframe from pandas.series
pd.DataFrame(y_train).head()  # print top 5 values
```

- price_range=0 means low pricw, 1=medium, 2=high, 3=very high as given on kaggle.com from where I've downloaded this dataset

```python
## convert to numpy arrrays for convenience
X_train = X_train.values
y_train = y_train.values
```

### Perform CHI square Test to get the relavance of X-features with y

```python
model = SelectKBest(score_func=chi2, k=10)
```

```python
model.fit(X_train,y_train)
```

```python
chi_scores = model.scores_
```

```python
## Now we can convert this scores in dataframe and add column names to all these values.
dfscores = pd.DataFrame(chi_scores)
dfcolumn = pd.DataFrame(data.columns[:-1])  # except last column which is for y_range in data

# concatenate 2columns columnwise by passing axis=1
feature_scores = pd.concat([dfcolumn, dfscores], axis=1)
feature_scores.columns = ["Features_name", "Relavance_score"]
feature_scores
```

```python
sorted_vals = feature_scores.sort_values(by="Relavance_score", ascending=False)
sorted_vals
```

```python
sorted_vals.values[:4,0]
```

```python
X_new_train = data[sorted_vals.values[:4,0]]
X_new_train.head()
```

## Feature Importance
- Feature Importance means calculating relavance of each feature to check the importance of each feature. 

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.feature_importances_
```

```python
np.sum(model.feature_importances_)
```

```python
featureImportance = pd.DataFrame(model.feature_importances_,index=data.columns[:-1],columns=["Importance"])
featureImportance.sort_values(by="Importance", ascending=False, inplace=True)
featureImportance
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(20,6))
plt.bar(featureImportance.index, featureImportance["Importance"])
plt.show()
```

- It shows how much each feature contributes and now we can take top k features from our data

## Check Model Performance
- We will check the model performance with all the features first and then we will check the model  performance with top k best features.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
rfc = RandomForestClassifier()
```

```python
# check score for X_train with all 20 features 
scores = cross_val_score(rfc, X_train, y_train, cv=10)
print(X_train.shape, scores, sep="\n")
```

```python
scores.mean()*100
```

##### Now calculate score for X_train with top k best features:

```python
# check score for X_new_train which we have modifier in CHI square method in begining  
scores = cross_val_score(rfc, X_new_train, y_train, cv=10)
print(X_new_train.shape, scores, sep="\n")
```

```python
scores.mean()*100
```

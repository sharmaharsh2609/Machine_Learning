## Locally Weighted Regression
1) Read and Normalise the Dataset 

2) Generate W for every query point!

3) No training is involved, directly make predictions using closed form solution we dervied!

`(X′WX)inv * X′WY` where X' is X_Transpose
- It's derivation is given in 05_close_sol_LocallyWeightedRegression

4) Find the best value of Tau(Bandwidth Parameter) [Cross Validation]

## 1. Loading Data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
dfx = pd.read_csv('../../Datasets/Locally Weighted Regression/weightedX.csv')
dfy = pd.read_csv('../../Datasets/Locally Weighted Regression/weightedY.csv')
```

```python
X = dfx.values
Y = dfy.values
```

```python
print(X.shape)
print(Y.shape)
```

```python
plt.style.use("seaborn")
```

```python
#Normalisze the Data

u = X.mean()
std = X.std()
X = (X-u)/std

plt.title("Normalized Data")
plt.scatter(X,Y)
plt.show()
```

- Now we have to predict theta for all the X_test points differently. There is no generic theta. That's why this is called non-parametric algo.

## 2. Find out W (mxm) diagonal matrix


```python
print(np.eye(5))
print(type(np.eye(5)))
```

```python
print(type(np.mat(np.eye(5))))

#a*b ==> a,b -->matrices -->Matrix Multicplication
#a*b ==>? a,b -->arrays --->Element Wise 
```

```python
# defining function for above equation
def getW(query_point,X,tau):
    M = X.shape[0]
    W = np.mat(np.eye(M))  # create a unit diagonal matrix.
    
    for i in range(M):  # find w for all m examples
        xi = X[i]
        x  = query_point
        W[i,i] =  np.exp(np.dot((xi-x),(xi-x).T)/(-2*tau*tau))
    return W
```

```python
X = np.mat(X)
Y = np.mat(Y)
M = X.shape[0]

W = getW(-1,X,1)
print(W.shape)
print(W)
```

```python
## Let us try it with passing high value of tau
W = getW(-1,X,100)
print(W.shape)
print(W)
```

### 3. Make Predictions


```python
def predict(X,Y,query_x,tau):
    ones = np.ones((M,1))
    X_ = np.hstack((X,ones))  # add x0 at last column to make it of size (m x n+1)
    
    # qx(query point) should be of same size as X_ . So we add 1 as last column to qx also 
    qx = np.mat([query_x,1])   # add 1 as last column to query point
    
    W = getW(qx,X_,tau)
    
    #theta = `(X′WX)inv * X′WY`
    theta = np.linalg.pinv(X_.T*(W*X_))*(X_.T*(W*Y))  # pinv means pseudo inverse.
    #print(theta.shape)
    pred = np.dot(qx,theta)  # predicted value= one_example_from_X * Theta_vector_for_that_X
    return theta,pred
    
```

```python
theta,pred = predict(X,Y,1.0,1.0)

print("Theta: ", theta)
print("Prediction: ", pred)
```

## 4. Visualise Predictions, Analyse the Effect of Tau

```python
# function to visualize the effect of tau
def plotPrediction(tau):
    X_test = np.linspace(-2,2,20)  # it generates 20 numbers in between interval of -2 to +2
    # we generated b/w -2 to +2 bcoz our data is normalized to mean=0 and std=1
    Y_test = []
    
    # predict theta for all the points inside X_test and append predicted result in Y_test
    for xq in X_test:
        theta,pred = predict(X,Y,xq,tau)
        Y_test.append(pred[0][0])
        
    Y_test = np.array(Y_test)  # convert to numpy array bcoz matplotlib do not accept np.matices
    # convert X and Y also to numpy arrays as:
    XO = np.array(X)  # XO and YO means X original and Y original
    YO = np.array(Y)
    
    plt.figure(figsize=(5,5))
    plt.title("Tau/Bandwidth Param %.2f"%tau)
    plt.scatter(XO,YO)
    plt.scatter(X_test,Y_test,color='red')
    plt.show()    
```

```python
plotPrediction(tau=1)
```

```python
taus = [0.1,0.5,1,5,10]
for t in taus:
    plotPrediction(t)
```

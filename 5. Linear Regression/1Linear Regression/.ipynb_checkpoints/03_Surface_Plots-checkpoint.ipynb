### Surface Plots | Data Visualisation

Surface Plots are used in many Machine Learning Algorithms. They are used to:
- Visualise Loss Functions in Machine Learning & Deep Learning
- Visualise State or State Value Functions in Reinforcement Learning

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from warnings import filterwarnings
filterwarnings("ignore") # to ignore all the warnings
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7])

x,y= np.meshgrid(a,b)
print(x)
print(y)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
z = x+y
axes.plot_surface(x,y,z,cmap='rainbow')
plt.show()
```

```python
plt.style.use('seaborn')
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(x,y,x**2+y**2,cmap='rainbow')
plt.show()
```

```python
x = np.arange(-1,1,0.02)  # from -1 to 1 with a gap of 0.02
y = x   # y is same as x
print(x.shape, y.shape)

x,y = np.meshgrid(x,y)
print(x.shape, y.shape)
```

```python
fig = plt.figure(figsize=(8,8))
axes = fig.gca(projection='3d')
axes.plot_surface(x,y,x**2+y**2,cmap='rainbow')
plt.show()
```

## Contour plot:
- If we cut above plot by parralel planes at different heights, then we get the cintour plots. See below.
- Here a plane will cut the bowl at same values. Thus the color of the whole ring will be same.

```python
fig = plt.figure(figsize=(7,7))
axes = fig.gca(projection='3d')   # create axes object for 3d axes
axes.contour(x,y,x**2+y**2,cmap='rainbow') 
plt.title("Contour Plot")
plt.show()
```

- These are the contours and these look similar if we cut our surface plot with parallel planes.

- Executing this code in python file, we can rotate and move the 3d space with our mouse/touchpad cursor.

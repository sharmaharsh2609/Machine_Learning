## Gradient Descent (in general)
- This method is used to give minimimum/maximum value of a function.
##### In general, this algorithm is used in a lot of ML algos including **deep learning, neural network etc**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)  # generate numbers from 0-9
Y = (X-5)**2        # generate another list y
print(X,Y)
```

**Goal:**

Given a Function f(x), we want to find the value of x that minimizes f, where f is the error function


### Visualisation
- Let us do some visualization first

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

```python
x = 0
lr = 0.1   # lr(learning rate) is a stepsize by which we will increment in each step

# take 50 steps in the downhill direction in the graph
for i in range(50):
    grad = 2*(x-5)   # gradient which is derivative of F(X) that is defined in a previous cell
    x = x - lr*grad  # decrease x each step by grad*lr as we are moving in downward direction
    y = (x-5)**2  # this is improved y in each step
    print(x)
```

- We can also store this y in each step in a list named 'error' and then we can shoe this erro r in each step. This error is the calculated y in each step 

```python
x = 0
lr = 0.1   # lr(learning rate) is a stepsize by which we will increment in each step

lst = []
plt.plot(X,Y)   # original X, Y
for i in range(50):
    grad = 2*(x-5)   
    x = x - lr*grad  
    y = (x-5)**2  
    lst.append(y)   # add y of each iteration in a list
    print(x)
    plt.scatter(x,y)  # scatter plot x,y for these 50 iterations
plt.show()
```

- It shows how our y approches towards the minimum value 0 at x=5

<hr>

- We can also see the lst array in which we stored all the iterations. It shows us how our error reduces in each iteration at what speed/rate

```python
# Plot the values of error

plt.plot(lst)
plt.show()
```

- Here we got almost 0 error after x= around 13. After that, it saturates

- Here wo got almost 0 error around x=13. But we did 50 iterations. So we can use a break statement as:
        if (delta(x)<=0.01):
            break
- If change in x is very small(here we took 0.01) in each iteration, then we stop

```python
x = 0
lr = 0.1   # lr(learning rate) is a stepsize by which we will increment in each step

x_prev = x # initialize x_prev with same val as x 
lst = []
plt.plot(X,Y)   # original X, Y
for i in range(50):
    grad = 2*(x-5)   
    x = x - lr*grad  
    y = (x-5)**2  
    lst.append(y)   # add y of each iteration in a list
    print(x)
    plt.scatter(x,y)  # scatter plot x,y for these 50 iterations

    delta_x = x-x_prev  
    x_prev = x
    if delta_x<=0.01:   # if change in 'x' is smaller than 0.01 then break out of loop
        break
plt.show()
```

```python
plt.plot(lst)
plt.show()
```

- Now we can see our value of x is 4.96 and it stops after that as change in x is smaller than 0.01 now. We can also change this to delta_x<=0.0001

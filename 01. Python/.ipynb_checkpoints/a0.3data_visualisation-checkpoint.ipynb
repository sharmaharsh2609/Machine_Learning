# Data Visualisation

matplotlib is a package used to visualize 2d graphics.

```python
import matplotlib.pyplot as plt
#we can also use command:    from matplotlib import pyplot as plt
#pyplot is a class of library matplotlib
```

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = x**2   #y is square of x
print(x,y)
```

```python
plt.plot(x,y)    #plots a graph with x-axis as <1st_arg> and y-axis as <2nd_arg>
#plt.plot(x,x**2)     this is same as      plt.plot(x,y)     where y=x**2
plt.show()       #to print the plot
```

```python
#above graph is not looking much good
# To check for what all styles are available, use comman:
plt.style.available
```

```python
#to use a style, use command:
plt.style.use('seaborn')
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,x**2,label="Stock price 1")    #label shows the label on the graph
plt.plot(x,x**2+3,label="Stock price 2")
plt.legend()   #we need to call this method to show label. else label is not shown even if we pass label arg in plt.plot


plt.title("My plot")   #to show title at top

plt.xlabel("Timestamp")    #To show label on x-axis
plt.ylabel("Price")        #To show label on y-axis

plt.show()
```

# Scatter plot

```python
plt.style.use('dark_background')   #example of another theme. seaborn is best. But you shoul try others also

plt.scatter(x,x**2,label="Stock price 1")    #shows scattered graph
plt.scatter(x,x**2+3,label="Stock price 2")
plt.legend()  

plt.title("My scatter plot")

plt.xlabel("Timestamp")    #To show label on x-axis
plt.ylabel("Price")        #To show label on y-axis

plt.show()
```

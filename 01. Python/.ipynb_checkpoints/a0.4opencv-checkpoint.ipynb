# OpenCV & Images  
OpenCV (Open Source Computer Vision) Library is an open source computer vision and machine learning software library.  
-opencv stores image in BGR format. But an image is stored in RGB format. So opencv stores R as B and b as R which makes wrong colors.  
-To make these right, there is a function cv2.cvtColor() which converts color format of an image

```python
import cv2          #to import opencv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('./img01.png')     #pass the path of image as argument.   
#cv.imread() returns a numpy array. Thus img1 is a numpy array and we can perform numpy functions on it.

print(img1)
#img1 stres data values of colors RGB
```

```python
img1.shape    # output shows that image has 503 x 773 resolution with 3 levels 
#      -here img1 array has 3lists where 1st one stores color values of Red for 503 x 773 pixels,
#       2nd stores color values for green and 3rd stores color values for blue 
```

```python
plt.imshow(img1)  #to show image using matplotlib.pyplot.imshow()
```

```python
cv2.imshow('a',img1)

cv2.waitKey()       #press any character and cv2.waitKey will close current window
cv2.destroyAllWindows()     # destroys all opened windows
```

```python
img2 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)  #there are many more arguments for different color formats to change and from 

plt.grid()    #it makes a grid on image
plt.imshow(img2)
#now it shows us the real image 
```

```python
plt.imshow(img2)
plt.axis("off")    # it turn off the numbers displayed on axis and we can turn it on again by passing "on" argument.
plt.show()
```

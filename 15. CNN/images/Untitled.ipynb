### Image Segmentation using Thresholding and Clipping method.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('conv_nn.png')  # cv2 reads image in BGR
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert from BGR to RGB
```

```python
plt.imshow(img)
```

```python
## This below is called image negative 

img2 = 255 - img
plt.imshow(img2)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY )
plt.imshow(gray_img, cmap='gray')
```

```python
gray_img.shape
```

```python
gray_img[0,0]
```

```python
gray_img = 255 - gray_img
plt.imshow(gray_img, cmap='gray')
```

```python
mean = gray_img.mean()
m,n = gray_img.shape
for i in range(m):
    for j in range(n):
        if gray_img[i,j] > mean :
            gray_img[i,j] = 255  # assign white color to identify object
plt.imshow(gray_img, cmap='gray')  
```

- This is one method of image segmentation.
- Another is clipping where we use condition: `if gray_img[i,j] > 50 and gray_img[i,j] < 150`
. This is when 2 objects have different intensities, then we can classify one of them.

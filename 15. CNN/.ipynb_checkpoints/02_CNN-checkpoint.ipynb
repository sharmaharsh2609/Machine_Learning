### What is Convolution and how does it work
In maths convolution of 2 function results in a 3rd finction. So here we apply a filter on image and we get convolved_feature as output which contains extracted features from the image depending on the filter.

<img src="img2.gif" align="left" alt="image not loaded" style="width: 450px;">
<br>

- Here filter is: $$\begin{bmatrix}1 & 0 & 1\\
0 & 1 & 0\\
1 & 0 & 1
\end{bmatrix}$$
<br>

- Multiplying this filter with given image gives us that convolved feature array
<br>

- **Output array has higher value at area where filter has high similarity with image**

<font color="cyan">If image_size=(n,n) and filter_size=(f,f), then Output_size=**(n-f+1,n-f+1)**</font>
- rg- In above ex, img=(5,5) and filter=(3,3). So convolved_feature=(5-3+1,5-3+1)=(3,3)

- Above was for 2d grayscale image. For 3d images, we have a volume (w,h,c) c-channels and convolution at that lvl is like this:

## Convolutional Layer?
- Convolutional Layer are made up of different kind of filters, where every filter represents some feature of the image.
- In above, we saw convolution of image and filter gives us a single scalar value. We got 9 scalars above. These all 9 scalars are convolved features. And a convolved feature will fire a neuron if it has high activation.

```python
def draw_img(img, title="Image"):   # takes img, title as input
    plt.figure(figsize=(6,4))
    plt.imshow(img, cmap="gray")  ## pass grayscale map, bcoz we need it earlier
    plt.axis("off")    
    plt.title(title+str(img.shape))  ## In title print title and shape of image
    plt.show()
```

```python
img_ = cv2.imread("sample_img.jpg")  # cv2 reads in BGR gormat
img_ = cv2.cvtColor(img_, cv2.COLOR_BGR2RGB) # convert to RGB
img = cv2.resize(img_, (100,100))

draw_img(img, title="Sample Image ")
```

- This is our Image. And In title, we have title and Image shape as we have defined in draw_img().

- For 2d images, 2d filters are required. But generally we have 3d images and for that we need 3d filters and output of convolution over the image after applying filter is also a 3d.
- If image size=(n,n,3) and filter=(f,f,3), then output=(n-f+1,n-f+1,3).
- **If colors of image are not important, we only need shape, then we can convert 3d image to a 2d grayscale image to save memory.** But in most of cases, color is a very imp. feature.
- For now, to understand we will use 2d images and filters. But in general, we will apply 3d filters later on. So we will use a grayscale image and 2d filter.

```python
img_ = cv2.imread("sample_img.jpg")  # cv2 reads in BGR gormat
img_ = cv2.cvtColor(img_, cv2.COLOR_BGR2GRAY) # convert to gray
gray_img = cv2.resize(img_, (100,100))

draw_img(gray_img, title="Sample Image ")
```

- We can now see, image is in grayscale. And it has 100,100 shape which is 2d now.
- Earlier it was not that much clear, so I passed cmap="gray" in plt.imshow(). Bcoz matplotlib was plotting according to RGB earlier and we were passing grayscale values which is only 2D values. And plt.imshow() was wxpecting 3d image. So we need to define cmap="gray"

## Applying Convolution now:
- In CNN, filters are automativally learned by the model to get the features.
- But for understanding, we will define wxternally hardcoded filters.
- There are a lot of filter in DIP and computer vision for different purposes like Blur filter to make img blur, egde filter to make edges more clearly visible, contrast filter to change contrast etc..

```python
def convolution(img, img_filter):
    W,H = img.shape   ## width and height of image
    
    f = img_filter.shape[0]  # considering a square filter
    new_img = np.zeros((W-f+1, H-f+1)) # Create array to store output image 
    ## try below code on copy pen
    ## 2 loops are to iterate over every row and col of output image
    for row in range(W-f+1):
        for col in range(H-f+1):
            # instead of using 2 more for loops, we can use direct vectorization like below:
            new_img[row,col] = np.sum(img[row:row+f,col:col+f]*img_filter) 
            # '*' means element wise multiplication and then we add all elements
## there may be a case when above result becomes >255 or <0. But python only takes value in
# range 0-255. So we can convert >255 to 255 and <0 to 0 as below:
            if new_img[row,col] > 255:
                new_img[row,col] = 255
            elif new_img[row,col] < 0:
                new_img[row,col] = 0
    
    return new_img # this is activation map.
```

```python
blur_filter = np.ones((3,3))/9.0  # this is a 3,3 blur filter
blur_filter
```

- This is a blur filter as it will decrease bcoz during elementwise multiplication, we mutiply every element with 0.11 and then take their sum. So total pixel value is decreased.
- Now let us apply convolution on 2d grayscale image using blur filter

```python
output1 = convolution(gray_img, blur_filter)
draw_img(gray_img)  ## original image
draw_img(output1)   ## image after appying blur filter
```

- See image is blurred after applying blur filter.
- **Also note that image size is now (98,98)**, which we got as (100-3+1,100-3+1)

```python
## let us see edge filter now:
edge_filter = np.array([[1,0,-1],  # this is a vertical edge filter in particular
                       [1,0,-1],
                       [1,0,-1]])  
output2 = convolution(gray_img, edge_filter)
draw_img(gray_img)  ## original image
draw_img(output2)   ## image after appying edge filter
```

- So now all the vertical edges are highlighted adter applying vertical egde filters.
- This filter increases intensity of edges and decreases intensity after the edges which makes it even more clear. We can see this from filter also:
        [ 1, 0, -1 
          1, 0, -1
          1, 0, -1]

<font color="cyan">**In general, filter_values are automatically predicted by convolutional layer. We do not need to define them explicitely**.</font> Filter in convolutional layer are like weights in dense layer. CNN automatically gets the filter values to get required features for predicting results.

- **These hardcoded filters were used in early days when neural networks were not into existence.**

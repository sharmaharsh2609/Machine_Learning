## Data Augmentation
- Data Augmentation is a technique to generate more data from our previous data by slightly changing angles and noise and other parameters. 

```python
img_path = "../Datasets/Images/images/horses/horse-16.jpg"
```

```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
# convert img to array. And lets take target size = 100,100
img = image.load_img(img_path,target_size=(100,100)) 
type(img)
```

```python
img
```

```python
x = image.img_to_array(img)/255.0  # normalize data
```

```python
plt.imshow(x)
plt.axis("off")
plt.show()
```

```python
x.shape
```

```python
x = x.reshape((1,100,100,3))
```

```python
datagen = image.ImageDataGenerator(
    rotation_range = 30,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    fill_mode = "nearest"
)
```

```python
type(datagen)
```

```python
i = 0
for batch in datagen.flow(x,batch_size=1):
    plt.figure()
    plt.axis("off")
    imgplot = plt.imshow(image.img_to_array(batch[0]))
    i+=1
    if i==10:
        break
    plt.show()    
```

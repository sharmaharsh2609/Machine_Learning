- In this example, we are given 5features of air and air quality index for each sample.
- We have to predict the air quality for X_test samples after training our algo for this.

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("./Train.csv")
df.head()
```

```python
X_train = df.values[:,:-1]  # all rows and all columns upto last one
y_train = df.values[:,-1]  # all rows and last column

X_train.shape, y_train.shape
```

- We have 1600 samples of air and 5 features for each sample and y represents the air quality index for all these 1600 examples.

```python
## using sklearn model
model = LinearRegression()  # create model object
model.fit(X_train, y_train) # fit the data
```

## Testing Part:

```python
X_test = pd.read_csv("./Test.csv")
X_test.head()
```

```python
X_test = X_test.values
y_pred = model.predict(X_test)
```

```python
## convert to csv file
df = pd.DataFrame(y_pred, columns=["target"])
df.to_csv("output.csv", index_label="ID")
```

### We can do same by using our own algo also:

```python
def hypothesis(X,theta):
    return np.dot(X,theta)

def error(X,y,theta):
    e = 0.0
    y_ = hypothesis(X,theta)
    e = np.sum((y-y_)**2)
    m = X.shape[0]
    return e/m
    
def gradient(X,y,theta):
    
    y_ = hypothesis(X,theta)
    grad = np.dot(X.T,(y_ - y))
    m = X.shape[0]
    return grad/m

def gradient_descent(X,y,learning_rate = 0.1,max_iters=300):
    
    n = X.shape[1]
    theta = np.zeros((n,))
    error_list = []
    
    for i in range(max_iters):
        e = error(X,y,theta)
        error_list.append(e)
        
        #Gradient descent
        grad = gradient(X,y,theta)
        theta = theta - learning_rate*grad
        
    return theta,error_list
```

```python
theta, error_list= gradient_descent(X_train, y_train)
```

```python
theta
```

```python
# visualize our error in each iteration
plt.plot(error_list)
plt.show()
```

- Error decreased from 12000 to around 400.

### Predictions

```python
y_predicted = hypothesis(X_test, theta) 
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred, y_predicted)
```

- Here y_pred is prediction made by sklearn and y_predicted is prediction made by our own algorithm. We can see prediction made by sklearn regression model and by our own algo matches upto 99.80%. This isn't the accuracy of the program. This is just how much our predictions match by predictions made by sklearn

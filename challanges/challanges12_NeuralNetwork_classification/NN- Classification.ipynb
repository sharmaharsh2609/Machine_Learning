### Data Preparation and Preprocessing

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
XT = pd.read_csv("./Training Data/Logistic_X_Train.csv")
XT.head()
```

```python
YT = pd.read_csv("./Training Data/Logistic_Y_Train.csv")
YT.head()
```

```python
plt.style.use('seaborn')
plt.scatter(XT['f1'], XT['f2'], c=YT.values, cmap=plt.cm.brg)
```

- This is our dataset.

```python
## convert X and Y dfs to numpy arrays 
X_train = np.asarray(XT.values).astype('float32')
Y_train = np.asarray(YT.values).astype('float32')
print(X_train.shape, Y_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# define model
model = Sequential()
model.add(Dense(8,activation='relu', input_shape=((2,))))
model.add(Dense(3,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
## compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
```

- Neural Networks are prone to overfitting. So we will create a validation set and do "early stopping" when validation accuracy starts decreasing.
- We can either split training data using sklearn.model_selection.train_test_split and pass `validation_data=(x_val,y_val)` or we can directly pass `validation_split=0.20` to take 20% of data from training data as validation set.

```python
## train
hist = model.fit(x=X_train,y=Y_train,batch_size=512, epochs=600, validation_split=0.2)
```

## Visualizing data now

```python
h = hist.history
h
```

```python
plt.plot(h['loss'], label="Training Loss")
plt.plot(h['val_loss'], label="Validation Loss")
plt.xlabel("No of epochs")
plt.ylabel("MSE Loss")
plt.legend()
```

```python
plt.plot(h['accuracy'], label="Training accuracy")
plt.plot(h['val_accuracy'], label="Validation accuracy")
plt.xlabel("No of epochs")
plt.ylabel("accuracy")
plt.legend()
```

```python
train_idx = np.argmax(h['accuracy'])
print("At index:", train_idx, "maximum training accuracy is: ",  h['accuracy'][train_idx])

val_idx = np.argmax(h['val_accuracy'])
print("At index:", val_idx, "maximum validation accuracy is: ", h['accuracy'][val_idx])
```

```python
X_test = pd.read_csv("./Test Cases/Logistic_X_Test.csv")
X_test = X_test.values
```

```python
predictions = model.predict(X_test)
outputs = np.zeros((len(predictions),1))
outputs[predictions>0.5] = 1
outputs
```

```python
df = pd.DataFrame(outputs, columns=['label'])
df.to_csv('Output.csv', index=False)
```

```python
import numpy as np, matplotlib.pyplot as plt

from sklearn.datasets import make_regression
```

```python
type(make_regression)
```

```python
X,y = make_regression()

print(X.shape, y.shape)
```

```python
X, y = make_regression(n_samples=1000, n_features=2, n_informative=1, noise=10, random_state=1)

X.shape, y.shape
```

```python
plt.style.use("seaborn")
```

```python
plt.subplot(1,2,1)  #no of rows, no of cols, number of the plot from all the plots
plt.scatter(X[:,0], y)

plt.subplot(1,2,2) 
plt.scatter(X[:,1], y)
plt.show()
```

##### Here y depends on 2features. We can visualize it in 3d plots using following code

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize = (10,7))
ax = plt.axes(projection="3d")

## Creating Plot
ax.scatter3D(X[:,0], X[:,1], y, color="green")
plt.title("Visualization linear regression Model for 2 features:")
plt.show()
```

## Implementation using sklearn model
- We can do all the training and testing for linear regression with help of sklearn model.
<br>https://scikit-learn.org/stable/modules/linear_model.html 
- If we open this link, all the instructions are given about how to train data, how to fit data, how to test data etc..

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
## We need to fit the data to the object first.
model.fit(X,y)
# fit() method tells about the data to all other methods, what data we are processing on
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
# male predictions for first 5elements of X
model.predict(X[:5])
```

```python
# we can check real y for first 5 elements also
y[:5]
```

```python
model.score(X, y)
```

- It takes X_test as 1st argument and y_test as 2nd argument.
<br>[As we didn't have test data, so we passed train data to this method.]

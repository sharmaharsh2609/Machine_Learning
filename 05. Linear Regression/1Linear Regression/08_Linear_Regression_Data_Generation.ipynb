```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.arange(20)
print(X)
```

```python
Theta = np.array([2,3])  # for generation, we know theta for ideak case

# now we add some noise in the ideal case. For linear regression our noise is "guassian noise".
# guassian noise is a normal distribution with "0-mean" and some "sigma-std"
# np.random.randn(20) create 20 sample points from standard normal distribution with 0-mean, 1-std
# we can convert it to any distribution by: "sigma*np.random.randn(20)+ u"
# In case of guassian noise, u=0. So we only multiply by some sigma
noise = 3*np.random.randn(20)
```

```python
y_ideal = Theta[1]*X + Theta[0]   # This is ideal equation of line
# This is some real life data, where we have some noise in the data
Y_real = Theta[1]*X + Theta[0] + noise
```

```python
plt.style.use("seaborn")
plt.plot(X,y_ideal,color='orange')
plt.scatter(X,Y_real)
plt.show()
```

- This is the ideal line with orange color and the scatter plot with blue color represents some real life regression data which has some noise in it.

### (Extra Topic)

- In Linear Regression, we update theta such that our loss/error should be minimum.
- To do this, we used Gradient Descent Algorithm. But there is one more algorithm named **Closed Form Solution** which can also be used to get the value of theta such that loss should be minimum.

### Closed Form Solution in Linear Regression
 $\theta = (X^TX)^{-1}X^T\hat{y} $
- This is the formula for Calculating Linear Regression using Closed Form Solution
- It's proof is derived in "10_closed_form_sol.pdf" file. In this file, this above formula is derived first and then differences in the closed Form and Gradient Descent are given

## Difference in GD and Closed Form Solution:
- GD is iterative method and it can become slow if learning_rate is very low. And closed form solution directly get the solution. But we only use closed form for small datasets.
- Closed form is very expensive for large dataset as $(X^TX)$ is of order (nxn) [bcoz (X=nxm) and $X^T$=mxn] Thus finding inverse of (nxn) $(X^TX)^{-1}$ can be very expensive for large value of n.
- As we generally have large dataset in ML, so using mini-batch GD is beneficial over closed form solution. That's why closed form solution is not used much.
- And if we have large data which can't fit into memory, then mini batch GD is better approach as it passes data in batches.
- Only advantage of closed form  solution is that we do not need to take care of mini batch. So **Most commonly we use mini-batch GD**.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
```

```python
# Generate Dataset
X,Y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)

print(X.shape, Y.shape)
```

```python
## Y is a vector of size:(400,).  Reshape Y into a matrix of size (400x1)
Y = Y.reshape((-1,1))
print(X.shape, Y.shape)
```

```python
# Normalize our data
X = (X-X.mean())/X.std()

# Visualize data now
plt.style.use("seaborn")
plt.figure(figsize = (5,5))
plt.scatter(X,Y)
plt.title("Normalized Data")
plt.show()
```

- Here we have only one feature. So we need theta = [ theta[0], theta[1] ]
- And we need to add array of ones. Only difference is: Here we add the array of ones at last. See the proof of algo in "10closed_form_sol.pdf" file.

```python
ones = np.ones((X.shape[0],1))
X_ = np.hstack((X,ones))  # stack ones array as last column 
print(X_.shape)
print(X_[:5,:])  # see first 5 examples
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def getThetaClosedForm(X,Y):    
    Y = np.mat(Y)  # if Y is a vector, then convert it to a matrix
    
    firstPart = np.dot(X.T,X)
    secondPart = np.dot(X.T,Y)
    # see formula above. This is implementation of that formula
    theta = np.linalg.pinv(firstPart)*secondPart
    return theta
```

```python
theta = getThetaClosedForm(X_,Y)
print(theta)
```

```python
## plot the predicted line wiith different color and plot scatter plot for the data
plt.figure()
plt.scatter(X,Y)
plt.plot(X,predict(X_,theta),color='red',label="prediction")
plt.title("Normalized Data")
plt.legend()
plt.show()
```

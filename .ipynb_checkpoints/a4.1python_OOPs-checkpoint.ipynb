## 1. Encapsulation: 
###### Encapsulation is the idea of wrapping up the data and the methods(functions) that work on data in a single unit.
- This puts restrictions on accessing variables and methods(functions in C++) directly and can prevent the accidental modification of data.
- To prevent accidental change, an object’s variable can only be changed by an object’s method. Those types of variables are known as private variables.
- As only class members can access data and no outer code can access a class data. It makes code more secure.
- one example of encapsulation is 'class' as class wraps it's variables(data) and member functions(methods) in a single unit.
- Another feature of encapsulation is that it makes easier to maintain the information as we can access data easily. This is defined as below image:

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("a4.1img.jpeg")

plt.figure(figsize = (15,10))
plt.imshow(img)
```

- We can see here using class, we just need to store address of object and we can access all variables, functions using objects address for earch record.
- But using fuctions, we need to access address of all variables, function differently which makes 2nd method a lot  difficult to maintain.

## 2. Abstraction: 
- Abstraction means only showing essential features and hiding the unneccessary information. Means only showing the results and hiding up the implementation.  
  e.g- We have a class exam{}; and we have answer_sheets(as objects). We execute evalute() function of class 'exam' for every answer_sheet(object). Here evaluate() takes input as answer sheet and shows result as marks. But evaluate() don't show us the procedure of evaluation that how it evaluates. evaluate() shows us the essential result and not the implementation. This is hiding unneccessary information from user.
- Abstraction dictates that user may not want to know the implementation of the functions. All user need to know is the signature. (A function(or method) signature defines the input and output of function.)  
 eg- Here above, evaluate(answer_sheet) and output combined are signature 

### Difference in Encapsulation and Abstraction:
- Data Abstraction and Data Encapsulation have many similarities, but they both are different concepts.
##### Data Abstraction is method of hiding unwanted information that user do not want to know.  Whereas encapsulation is a method to hide the data in a single entity or unit along with methods to protect information from outside.
- In abstraction, complex implementation  are hidden using abstract classes and interfaces. While in encapsulation, the data is hidden using methods of getters and setters. The objects that help to perform abstraction are encapsulated. Whereas the objects that result in encapsulation need not be abstracted.
- Thus encapsulation is neccessary for anstraction which explains why "abstraction is implemented through encapsulation". However encapsulation is independent and don't need abstraction.
- There are many more differences. But these are the major ones.

## 3. Inheritance:
- Inheritance is the capability of one class to inherit the properties from another class or to derive properties to other class. The class that derives properties is called the derived class or child class and the class from which the properties are being derived is called the base class or parent class.
- Inheritance provides re-uability of code.
- Inheritance allows us to define a class in terms of another class, which makes it easier to create and maintain an application.
##### Need of Inheritance: 
- Consider we have some code C1. Now we need another code C2 in same program which have similar properties as the code C1, then we can inherit C2 from C1 and then add more features in C2 which were required. Here without the concept of Inheritance, we need to write all the same code for C2 again. But using inheritance makes it short.(This is example of code re-usability).
- another use is when 2 or more classes have a lot of same code, then we can use a base class which contains the common code and then both the classes will be derived from the base class and then add their uncommon part in their own classes.
eg- shape is base class. It contains 2attributes: height and base. Then rectangle and triangle are derived from shape which both have different functions to implement .

- by polymorphism, we mean we can perform a task in different ways.  
eg- In C++, we can use function overloading, which is an example of polymorphism.

### self:
- Let us take example of a class:
        class Person():
            def __init__(self,name):  
                self.name1 = name
            def show(self):
                print("My name is ", self.name1)
            
        p1 = Person("Harsh")
        p2 = Person("Adarsh")
        p2.show()
            
- Here in above example we havae 2 objects of class Person(): p1 and p2.
- When we call p2.show() , then control directly goes to def show() in class Person. But compiler don't know which object is calling this show(): whether p1 is calling show or p2 as class methods(or functions) use object's attributes. eg- here show() method uses name1 attrobute. So it need to know which object's attributes it should access.
##### In python, whenever a method is called, python interpreter automatically passes object name as it's first argument, so that 'method'(or member function) would understand which object is calling that method. This is called as self.(self is not a keyword. It is a general name given for passing object itself as 1st argument). 
- self is not a keyword. any identifier as 1st argument in a function contains address of object.
- In C++, it is done using 'this' pointer. 'this' pointer is a hidden pointer which during compilation, stores the base address of the object.

```python
class Person(): 
        name = "Harsh"
        def show():
            print("My name is ", name)

p1 = Person()
p1.show()
```

- See, here we do not pass any argument, but it still shows us that 1 argument was passed. If we pass 1 argument, then it will show 2 arguments were passed. This is because whenever we call a method in python, 1st argument is automatically passed by interpreter as 'name of object' calling that method.
- So we always need to pass 1 argument less to an object method, because 1st one is ;automatically passed object name'.

```python
class Person(): 
        name = "Harsh"
        # we can write any identifier name in place of 'self'. But it is good practice to use self
        def show(self):    
            print("My name is ", self.name)

p1 = Person()
p1.show()

#In python, we can call functions using abstract classes directly by passing object name as:
Person.show(p1)
```

- Here we passed 0 argument. But we define show() with one argument which contains the object name that called the method.
- Also we can't directly use attributes of class. For example we can't use name directly here. That will show error : "name is not defined". Try this out. We need to write self.name [

### \__init__ method:
- python has some special methods: They start and end with double underscore. These are called dunders or magic functions.
- One of the special method in python is: \__init__ method.
- init method is called as soon as an object is instantiated.
##### \__init__ method is just like constructor. But __init__ method is strictly not a contructor.
- We can use \__init__ method to initiate the values of object's attributes.
#### Difference in \__init__ method and constructors in c++:
- Both \__init__ methof and constructors are used to initiate values of the attributes of an object.
- However main difference in both is: **when new object is created in python, object has no attributes before executing \__init__ method. Attributes are created in the execution of \__init__ method.   
 But in C++ when an object is created, attributes inside constructor are all pre defined, before the execution of constructor, but they store some garbage value before running constructor.**
 #### Use of \__init__ method: 
 - we can use \__init__ method to initialize the values of attributes for an object.


```python
#example to show __init__
class Person():
    def __init__(self):
        print("New object created")

p = Person()
print("Hey")

#we can see, __init__ method is executed as soon as object is created. then after it gooes to next statement
```

```python
#use of __init__
# as class is an callable object and we can use paranthesis on a callable object.
class Person():       # writing paranthesis is optional and are generally avoided
    def __init__(self):    # it creates and initializes attributes name, branch, college for every obj.
        self.name = "Harsh"
        self.branch = "CSE"
        self.college = "NITH"
    def show(self):
        print("My name is ", self.name)

p1 = Person()
p1.show()
```
